```python
import numpy as np
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
df = h2o.import_file("water_potability.csv")
```

```python
df_train, df_test = df.split_frame(ratios=[0.8])
```

```python
X = df.columns
X.remove("Potability")

y = "Potability"
```

```python
h2o_ml = H2OAutoML(max_runtime_secs=300, max_models=10, seed=10, verbosity="info", nfolds=2)
```

```python
h2o_ml.train(x=X, y=y, training_frame=df_train)
```

```python
models = h2o_ml.leaderboard
```

```python
models
```

```python
model = h2o.get_model("StackedEnsemble_BestOfFamily_1_AutoML_1_20230516_165233")
model.params
```

```python
y_pred = model.predict(df_test)
```
